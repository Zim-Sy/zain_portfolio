# Analysis of directorial debuts
## Data Cleaning Phase

```python
import pandas as pd

pathname = '/Users/zainsyed/Desktop/Pandas_working with files/Pandas Project files/imdb_movies.csv'
movies = pd.read_csv(pathname)

# Displaying All Columns
pd.set_option('display.max.columns', 22)

# dropping duplicate row
movies.drop_duplicates(inplace = True)

# Dropping any rows that contain null values found in Genres column
movies.dropna(subset = ['genres'], inplace = True)

# Creating 'Profit' column
movies['profit'] = movies['revenue'] - movies['budget']

# Dropping Columns we don't need 
# Note: (doing this will reduce loading time and make analysis more manageable)
movie_genres = movies.drop(['id', 'imdb_id', 'cast', 'homepage',
                            'tagline', 'keywords', 'overview',
                            'production_companies', 'budget_adj','revenue_adj',
                            'runtime'],
                            axis = 1)
```

```python
# Cleaning/Re-shaping 'genres' column
from pandas import Series, DataFrame

split = movie_genres['genres'].str.split('|').apply(Series, 1).stack()
split
# index of 'split' is 1 row ahead...
# dropping index of 'split' by 1
split.index = split.index.droplevel(-1)
split

# Renaming split
split.name = 'genres_split'

# Replacing genre column with reshaped column
del movie_genres['genres'] 
movie_genres = movie_genres.join(split)
```

```python
# Re-formatting release_date
movie_genres['release_date'] = pd.to_datetime(movie_genres['release_date'])
movie_genres.head()
```

```python
# Reordering movies by release_year
movie_genres.sort_values(by = 'release_date', ascending = True, inplace = True)
```

```python
movie_genres[movie_genres['release_date'].dt.year >= 2016].sort_values(by = 'release_date', ascending = True)
```

```python
mask = (movie_genres['release_date'].dt.year.between(2060, 2074)) & \
       (movie_genres['release_date'].dt.year - movie_genres['release_year'] == 100)

movie_genres.loc[mask, 'release_date'] = movie_genres.loc[mask, 'release_date'] - pd.DateOffset(years=100)
```

```python
movie_genres.sort_values(by = ['release_date','director'], inplace = True)
movie_genres
```

```python
movie_genres.dropna(subset = ['director'], inplace = True)
```

```python
movie_genres[movie_genres['director'].str.contains('|', regex = False)]
```

```python
# Split directors into lists
movie_genres['director'] = movie_genres['director'].str.split('|')  

# Explode lists into separate rows
movie_genres = movie_genres.explode('director', ignore_index = True)  
```

```python
movie_genres.sort_values(by = ['release_date','director'], ignore_index = True, inplace = True)
```

```python
movie_genres.tail(7)
```

```python
movie_genres[movie_genres['director'].str.contains('Goetz')]
```

```python
movie_genres.drop_duplicates('director', keep = 'first')
```

```python
debut_directors = movie_genres.groupby('director')['release_date'].min().reset_index()
debut_directors.rename(columns = {'release_date': 'debut_date'}, inplace = True)

# Merge debut information back into the original DataFrame
movie_genres = movie_genres.merge(debut_directors, on = 'director')

# Filter rows where the release_date matches the director's debut_date
movie_genres = movie_genres[movie_genres['release_date'] == movie_genres['debut_date']].drop(columns='debut_date')
```

```python
movie_genres
```

```python
movie_genres[movie_genres['director'].str.contains('Goetz')]
```
